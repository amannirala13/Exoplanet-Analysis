```python
pixelfile.plot()
```

```python
# Converting the pixel data to light curve graph
lc = pixelfile.to_lightcurve()
lc.plot()
```

```python
# Flatten the graph for better understanding
flat_lc = lc.remove_nans().flatten()
flat_lc.scatter()
```

```python
'''
0.3 = minimum period value
4 = maximum period value
0.001 = interval between periods
'''
periodogram = flat_lc.to_periodogram(method="bls")
periodogram.plot()
period = periodogram.period_at_max_power
print("Period [days]:",period)
```

```python
# phase folding the signal frequency 
folded_lc = flat_lc.fold(period=period)
folded_lc.scatter()
```

```python
folded_lc = folded_lc.bin(binsize=10)
folded_lc.plot()
```
